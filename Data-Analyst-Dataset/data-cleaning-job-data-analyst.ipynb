# Import libraries and packages

```python
import numpy as np 
import pandas as pd 
```

# Load data

```python
data = pd.read_csv('../input/data-analyst-jobs/DataAnalyst.csv')
```

# Data Cleaning  

```python
data.head() #to get several line from top of data
```

```python
data.tail() #to get several line from end of data
```

remove Unnamed: 0 column

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.head()
```

Change name of columns

```python
data = data.rename(
    columns={'Job Title': 'Job_Title',
             'Salary Estimate': 'Salary_Estimate',
             'Job Description': 'Job_Description',
             'Company Name': 'Company_Name',
             'Type of Ownership': 'Ownership',
             'Easy Apply': 'Easy_Apply'})
```

check null value

```python
data.isnull().sum()
```

replace -1, '-1' and -1.0 as np.nan (keyword none)

```python
data.replace(-1, np.nan, inplace = True)
data.replace('-1', np.nan, inplace = True)
data.replace(-1.0, np.nan, inplace = True)
```

```python
data.isnull().sum()
```

In this section we know that there are too many "null" values in Competitors and Easy_Apply columns

```python
data = data.drop('Competitors', axis = 1)
```

```python
data = data.drop('Easy_Apply', axis = 1)
```
